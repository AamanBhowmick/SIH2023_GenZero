```python
#split dataset
from sklearn.model_selection import train_test_split
X = data[['abdominal_pain', 'acidity', 'back_pain', 'bladder_discomfort', 'breathlessness', 'brittle_nails', 'bruising', 'chest_pain', 'chills', 'cold_hands_and_feets', 'congestion', 'constipation', 'continuous_sneezing', 'cough', 'dark_urine', 'dehydration', 'diarrhoea', 'dizziness', 'excessive_hunger', 'fast_heart_rate', 'fatigue', 'headache', 'high_fever', 'hip_joint_pain', 'indigestion', 'itching', 'joint_pain', 'knee_pain', 'loss_of_appetite', 'loss_of_smell', 'muscle_pain', 'nausea', 'neck_pain', 'palpitations', 'polyuria', 'redness_of_eyes', 'runny_nose', 'shivering', 'spotting_ urination', 'stomach_pain', 'sweating', 'throat_irritation', 'ulcers_on_tongue', 'vomiting', 'watering_from_eyes', 'weakness_in_limbs', 'yellowing_of_eyes', 'yellowish_skin', 'blurred_and_distorted_vision', 'pain_in_anal_region']]
y = data['prognosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,
random_state=101)
```

```python
#train decision tree model
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
print("Decision Tree Classifier Created")
#predict on test data and calculate accuracy score
predictions = dtree.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```

```python
#visualize the graph of our decision tree classifier
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydot
features = list(data.columns[1:])
features
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,feature_names=features,filled=True,rounded=True)
graph = pydot.graph_from_dot_data(dot_data.getvalue())
Image(graph[0].create_png())
```
